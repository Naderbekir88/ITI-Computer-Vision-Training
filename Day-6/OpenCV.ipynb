###  Bind Trackbar To OpenCV Windows

```python
img = cv2.imread('turk.jpeg')
cv2.namedWindow('Blur')

def nothing(x):
    print(x)

cv2.createTrackbar('kSize','Blur', 1, 50 , nothing)

while (1):
    kSize = cv2.getTrackbarPos('kSize','Blur')
    blur = cv2.blur(img,(kSize,kSize))
    
    cv2.imshow('Blur',blur)
    k = cv2.waitKey(1)
    if k ==27:
        break
cv2.destroyAllWindows()
```

```python
def bright(img,beta_value):
    img_bright = cv2.convertScaleAbs(img,beta=beta_value)
    return img_bright
```

```python
img = cv2.imread('turk.jpeg')
bright_img = bright(img,70)
cv2.imshow('bright',bright_img)

# Instantiate the function
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

img = cv2.imread('turk.jpeg')
sobel_y = np.array(
    [
    [-1,-1,-1],
    [0,0,0],
    [1,1,1]
    ]
    ,np.float32)
out =cv2.filter2D(img,-1,sobel_y,borderType=cv2.BORDER_ISOLATED)
cv2.imshow('sobel_y',out)

# Instantiate the function
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

img = cv2.imread('turk.jpeg')
sobel_x = np.array(
    [
    [1,0,1],
    [-1,0,-1],
    [-1,0,1]
    ]
    ,np.float32)

sobel =sobel_x +sobel_y
out =cv2.filter2D(img,-1,sobel,borderType=cv2.BORDER_ISOLATED)
cv2.imshow('sobel_x',out)

# Instantiate the function
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

img = cv2.imread('cards.jpg')

def sharpen(img):
    kernal = np.array([[0,0,0],[0,1,0],[0,0,0]])
    img_sharpen = cv2.filter2D(img,-1,kernal)
    return img_sharpen
```

```python
out = sharpen(img)
cv2.imshow('Sharpen_image',out)

# Instantiate the function
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def HDR(img):
    hdr = cv2.detailEnhance(img,sigma_s=12,sigma_r=0.15)
    return hdr
#sigma_s Range :   0:200  Details of an img
#sigma_r 0:1
```

```python
out = HDR(img)
cv2.imshow('HDR',out)

# Instantiate the function
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def invert (img):
    inv = cv2.bitwise_not(img)
    return inv
```

```python
out = invert(img)
cv2.imshow('inverter',out)
# Instantiate the function
cv2.waitKey(0)
cv2.destroyAllWindows()
```
