```python
import os
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import cv2
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
tf.config.list_physical_devices('GPU')
```

```python
data_dir = 'cats_and_dogs/'
train_dir = os.path.join(data_dir,'train')
valid_dir = os.path.join(data_dir,'validation')
print(train_dir , valid_dir)

cats_train = os.path.join(train_dir ,'cats')
cats_valid = os.path.join(train_dir ,'cats')

dogs_train = os.path.join(valid_dir ,'dogs')
dogs_valid = os.path.join(valid_dir ,'dogs')

print(dogs_train)
```

```python
print(cats_train)
print(len(os.listdir(cats_train)))


print(cats_valid)
print(len(os.listdir(cats_valid)))

print(dogs_train)
print(len(os.listdir(dogs_train)))

print(dogs_valid)
print(len(os.listdir(dogs_valid)))
```

```python
batch_size = 128
img_shape  = 150
```

`The tight_layout`: is used to automatically adjust subplot parameters to give specified padding

```python
def plotImage(image_arr):
    fig , axes = plt.subplots(1,5,figsize=(20,20))
    axes =axes.flatten()    #collapsed into one dimension
    for img,ax in zip (image_arr,axes):
        ax.imshow(img)
    plt.tight_layout() 
    plt.show()   
```

```python
#horizontal_flip Randomly flipped True or false
image_generator = ImageDataGenerator(rescale=1/255 ,horizontal_flip=True)

# load data fro m directory
train_data_gen = image_generator.flow_from_directory(batch_size=batch_size,
                                                     directory=train_dir,
                                                     shuffle=True,
                                                     target_size=(img_shape,img_shape))

print(train_data_gen)
```

```python
augmented_images = [train_data_gen[0][0][0] for i in range(5)] ;# class , Batch , first img
plotImage(augmented_images)
```

## Rotate

`rotation_range=70` means randomly rotate between [-70,70]

```python
image_generator = ImageDataGenerator(rescale=1./255,rotation_range=70)
train_data_gen  = image_generator.flow_from_directory(batch_size=batch_size,
                                                     directory=train_dir,
                                                     shuffle=True,
                                                     target_size=(img_shape,img_shape))
```

```python
augmented_images = [train_data_gen[0][0][0] for i in range(5)] ;# class , Batch , first img
plotImage(augmented_images)
```

## Zoom

```python
image_generator = ImageDataGenerator(rescale=1./255,zoom_range=0.5)
train_data_gen  = image_generator.flow_from_directory(batch_size=batch_size,
                                                     directory=train_dir,
                                                     shuffle=True,
                                                     target_size=(img_shape,img_shape))
```

```python
augmented_images = [train_data_gen[0][0][0] for i in range(5)] ;# class , Batch , first img
plotImage(augmented_images)
```

```python
image_generator = ImageDataGenerator(rescale=1./255,horizontal_flip=True,
                                     rotation_range=40,
                                     zoom_range=0.2,
                                     height_shift_range=0.2,
                                     width_shift_range=0.2,
                                     shear_range=0.2,
                                    fill_mode='nearest')
train_data_gen  = image_generator.flow_from_directory(batch_size=batch_size,
                                                     directory=train_dir,
                                                     shuffle=True,
                                                     target_size=(img_shape,img_shape),
                                                     class_mode='binary')
```

```python
augmented_images = [train_data_gen[0][0][0] for i in range(5)] ;# class , Batch , first img
plotImage(augmented_images)
```

```python
image_gen_val =ImageDataGenerator(rescale=1./255)
valid_gen = image_gen_val.flow_from_directory(batch_size=batch_size,
                                            directory=valid_dir,
                                            target_size=(img_shape,img_shape),
                                              class_mode='binary')
```

## Model Creation

```python
model = tf.keras.Sequential([
   
    tf.keras.layers.Conv2D(32,(3,3), activation='relu',input_shape=(150,150,3)),
    tf.keras.layers.MaxPooling2D(2,2),
    
    tf.keras.layers.Conv2D(64,(3,3),activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    
    tf.keras.layers.Flatten(),
     tf.keras.layers.Dense(128,activation='relu'),
    tf.keras.layers.Dense(2)
])
```

```python
model.compile(optimizer='adam',
             loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics = ['accuracy']            
             )
```

```python
model.summary()
```

```python
EPOCHS = 20
history = model.fit(
    train_data_gen,
    steps_per_epoch=int(np.ceil(2000/float(batch_size))),
    epochs=EPOCHS,
    validation_data=valid_gen,
    validation_steps=int(np.ceil(1000/float(batch_size)))
    
        )
```

```python
# save model
model.save('final_model.h5')
```
